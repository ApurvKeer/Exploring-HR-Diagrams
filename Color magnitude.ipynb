```python
hd = []

with open(r'D:\IIT material\Krittika python project\hlsp_hugs_hst_wfc3-uvis-acs-wfc_ngc2808_multi_v1_catalog-meth1.txt', 'r') as fr:
        # reading line by line
        lines = fr.readlines() 
hd.append("Nan") 
hd.append("X")  
hd.append("Y")
        
for i in range(14, 45):
    hd.append(lines[i])
    
hd.append(lines[45])
hd.append(lines[46])
hd.append(lines[47])
hd.append(lines[48])
hd.append(lines[49])

print(hd)
```

```python
import pandas as pd
import numpy

file_path = r'D:\IIT material\Krittika python project\NGC2080 data.txt'

df = pd.read_csv(file_path, sep = ' ')

df.columns = hd

df.drop(['Nan'], inplace=True, axis=1)

df.drop(df[df['# Col. 33: Membership Probability [-1.0: not available]\n'] < 90].index, inplace = True)
df.drop(df[df['# Col.  3: F275W calibrated magnitude\n'] < 0].index, inplace = True)
df.drop(df[df['# Col.  9: F336W calibrated magnitude\n'] < 0].index, inplace = True)
df['Color index'] = df['# Col.  3: F275W calibrated magnitude\n'] - df['# Col.  9: F336W calibrated magnitude\n']

df
```

```python
import matplotlib.pyplot as plt

font = {'family': 'serif',
        'color':  'black',
        'weight': 'ultralight',
        'size': 20,
        }

x = df['Color index']
y = df['# Col.  3: F275W calibrated magnitude\n']

plt.figure(figsize = (15, 15))
plt.scatter(x, y, s = 0.0005, color = 'black')
plt.axis('square')
plt.gca().invert_yaxis()
plt.xlim(-2, 6)
plt.ylim(28, 16)
plt.title("NGC2808", fontsize = 30, y = 1.01, fontdict = font)
plt.xlabel(r'$m_{F275W} - m_{F336W}$', fontsize = 20, fontdict = font)
plt.ylabel(r'$m_{F275W}$', fontsize = 20, fontdict = font)
plt.tick_params(labelsize=15, pad = 6)
#plt.save_fig("D:\IIT material\Krittika python project\ColorMag.jpg")
plt.show()
```
